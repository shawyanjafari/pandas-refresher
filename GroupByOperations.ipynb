```python
import pandas as pd
df = pd.read_csv('data/city_weather_data.csv', parse_dates=['day'])
# df.set_index('day')
```

```python
# Create a group first
# in the args, you pass the name of the column you want to group by
grp = df.groupby('city')  # grp is of type DataFrameGroupBy
grp
```

```python
# each group name points to it's respective df
# austin is a key for the austin group df
# dallas is a key for the dallas group df, etc...
for city, city_df in grp:
    print(city)
    print(city_df)
```

```python
# get a specific group
grp.get_group('austin')
```

```python
# get all the analytics in one shot
grp.describe()
```

```python
grp.max() # returns max for each column for each group
```

```python
# get max temperature from a specific column
grp['day'].max()
```

```python
df = pd.DataFrame({
    'country': ['us', 'us', 'mex', 'mex', 'mex', 'jp', 'jp', 'ch'],
    'values_2': [1, 2, 3, 4, 5, 6, 7, 8],
})
df
```

```python
my_group = df.groupby('country')
my_group.size()
```

```python
# you can iterate through each group
for name, group in my_group:
    print(name, group['values_2'].sum())
```

```python
# Each group is it's own DataFrame
df = pd.DataFrame({
    'food': ['us', 'us', 'us', 'us', 'jp', 'jp', 'jp', 'jp'],
    'prices': [2, 2, 3, 1, 1, 4, 1, 2],
    ''
})
df
```

```python
# get max and min of prices for each country
min_and_max = df.groupby('country')['prices'].agg(['max','min'])
print(type(min_and_max))
min_and_max
```

```python
min_and_max.sort_values(by='max', ascending=True)
```
